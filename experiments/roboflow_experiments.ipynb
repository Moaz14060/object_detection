# <center>First Experiment<center>
---

```python
# !pip install ultralytics
from ultralytics import YOLO
```

##### **Without Undersampling**

```python
DATA_PATH = "F:/data.yaml"  

# Quick check: open data.yaml file
with open(DATA_PATH, "r") as f:
    print(f.read())
```

```python
# To check if the GPU is used intead of the CPU
import torch
print("CUDA available:", torch.cuda.is_available())
print("GPU Name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

```python
# Training
# Options: yolov8n.pt (nano), yolov8s.pt (small), yolov8m.pt (medium)
model = YOLO("yolov8n.pt")

results = model.train(
    data=DATA_PATH,   
    epochs=50,       
    imgsz=640,        
    batch=16,         
    workers=2         
)
```

```python
# Runs validation on test set
metrics = model.val()
print("Validation Results:", metrics)
```

```python
# Predection
# Test on a few sample images
results = model.predict(source="C:/Users/User/Desktop/istockphoto-478525372-612x612.jpg", save=True, imgsz=640, conf=0.25)
```

##### **Training a more stronger model**

```python
# Options: yolov8n.pt (nano), yolov8s.pt (small), yolov8m.pt (medium)
model = YOLO("yolov8s.pt")

results = model.train(
    data=DATA_PATH,   
    epochs=100,       
    imgsz=640,        
    batch=4,
    device=0,          
    augment=True
)
```

```python
model = YOLO("F:/runs/detect/train7/weights/last.pt")
model.train(resume = True)
```

```python
# Evaluation
metrics = model.val(data="data.yaml", split="test")
print(metrics)
```
